```python
def ultimate_predict(periods, history_forecast_ratio,df_orig, offset=8640):
    df = df_orig.iloc[-(offset+(int(60*periods*history_forecast_ratio))):-offset]
    PRICE_COLUMN = "usdtfutures_{}USDT_close_price".format(COIN)
    input_seq = df[PRICE_COLUMN].values
    d = {'ds': df.index, 'y': input_seq}

    df2 = pd.DataFrame(data=d)
    df2['ds'] = pd.to_datetime(df2['ds']).dt.tz_localize(None)
    
    m = Prophet()
    m.fit(df2)
    
    future = m.make_future_dataframe(periods=periods, freq='H')
    forecast = m.predict(future)

    correct_price = df_orig.loc[forecast.iloc[-1].ds][PRICE_COLUMN]
    forecast_price = forecast.iloc[-1].yhat
    hit = 0
    deciding_forecast = forecast_price
    #deciding_forecast = forecast.iloc[-1].trend
    
    if input_seq[-1] < correct_price and input_seq[-1] < forecast_price:
        hit =1
    if input_seq[-1] > correct_price and input_seq[-1] > forecast_price:
        hit =1

    side = 1
    if input_seq[-1] > forecast_price:
        side = -1
    
    

    return abs(abs(correct_price) - abs(forecast_price)), correct_price, forecast_price, input_seq[-1], hit, side
ultimate_predict(1, 2 ,df_orig, offset=len(df_orig)-15)
```

```python

p = {}

iter_results = {}

initial_bank = 400000
rake = 0.00018

iters = 500
for periods, hist_forecast_ration in [[48,0.5]]:
    errors = []
    hits = 0
    i = 0

    last_side = None
    for one in range(iters):
        i +=1
        offset_mult = len(df_orig)/(iters+2)
        offset_mult = int(offset_mult)# - np.random.randint(1,int(offset_mult)/5)
        offset = (one+1)*offset_mult
        offset = len(df_orig) - (40+i)
        res = ultimate_predict(periods, hist_forecast_ration ,df_orig, offset=offset)
        base_price = res[3]
        future_price = res[1]
        move = (future_price/base_price) - 1

        hit = res[4]
        side = res[5]
        if last_side is None:
            last_side = side
        profit = 1 if hit else -1
        change = initial_bank * abs(move) * profit
        initial_bank = initial_bank + change

        if last_side != side:
            rake_change = rake * initial_bank
            rake_change = rake_change * 2
            initial_bank = initial_bank - rake_change

        hits += hit
        error = res[0]
        errors.append(error)
        last_side=side
        print(initial_bank)
    
    iter_results[periods]={'acc':hits/i, 'error':np.mean(errors)}
```
