```python
import pandas as pd
import numpy as np
from xgboost import XGBClassifier, XGBRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
df_orig = pd.read_hdf("v2_apr21a.hdf")
#df_orig = df_orig[:int(len(df_orig)/2)]
df_orig=df_orig.fillna(0)
df_orig
```

```python
X = df_orig.loc[:,"usdtfutures_ADAUSDT_close_price_1_ago":]
X
```

```python
df = df_orig.copy()
cname = "y_usdtfutures_ADAUSDT_close_price"
Y_REAL = df[[cname]]
Y_MEAN = np.mean(np.abs(Y_REAL.values))
Y_MEAN = Y_MEAN
Y_MEAN
```

```python
Y_REAL
```

```python
Y_REAL = df[[cname]]
df.loc[df[cname] > Y_MEAN,cname]=3
df.loc[df[cname] < -Y_MEAN,cname]=2
df.loc[df[cname] < 2,cname]=1
Y_REAL = df[[cname]]
Y_REAL
```

```python
Y = Y_REAL
Y
```

```python
sum(Y["y_usdtfutures_ADAUSDT_close_price"])/len(Y["y_usdtfutures_ADAUSDT_close_price"])
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=6, shuffle=False)
```

```python
model = XGBClassifier()
model.fit(X_train, Y_train)
print(model)
```

```python
model.feature_importances_
```

```python
y_pred = model.predict(X_test)
predictions = [round(value) for value in y_pred]
```

```python
accuracy = accuracy_score(Y_test, predictions)
print("Accuracy: %.2f%%" % (accuracy * 100.0))
```

```python
initial_bank = 400000
res = []
rake = 0.00018
last_side = -1
for i in range(1,len(Y_REAL_ORIG),60):
    res.append(initial_bank)
    ypred = y_pred[i]
    if ypred ==1:
        continue
    side = -1
    if ypred ==3:
        side = 1
    
    profit = -1
    y_real = Y_REAL_ORIG[i][0]
    if side * y_real >0:
        profit = 1
    
    change = abs(y_real) * initial_bank * profit
    
                        
    initial_bank = initial_bank + change
    
    if last_side != side:
        rake_change = rake * initial_bank
        rake_change = rake_change * 2
        initial_bank = initial_bank - rake_change
    last_side = side
    

initial_bank, len(res)        
```
