```python
import pandas as pd
import numpy as np
PICKED_COIN = 'BTC'
#PICKED_COIN = 'BNB'
PICKED_COIN = 'ADA'
```

```python
#df = pd.read_hdf("apr21a.hdf")
df = pd.read_hdf("v2_lagging_may_6.hdf")
df = df.fillna(0)
#df = pd.read_csv("encoded.csv")
#df = df.drop(columns=['Unnamed: 0'])
#df = df.iloc[:int(len(df)/4)]
#df = df[:-180]
df
```

```python
enc_features = []
for one in df.columns:
    if 'enc_feature' in one:
        enc_features.append(one)
enc_features
```

```python
trade_in_columns = []
for one in df.columns:
    if "close_price" in one:
        trade_in_columns.append(one)
    else:
        break
trade_in_columns

```

```python
rest_features = []
for one in df.columns:
    if one in trade_in_columns:
        continue
    if one in enc_features:
        continue
    rest_features.append(one)
```

```python
currencies = []
for one in trade_in_columns:
    _one = one.split('_')[-3]
    currencies.append(_one)
currencies = list(set(currencies))
```

```python
first_cols = []
cols_to_remove = []

for curr in currencies:
    #df['final_trade_{}'.format(curr)] = df['trade_in_3h_usdtfutures_{}_close_price'.format(curr)] + df['trade_in_2h_usdtfutures_{}_close_price'.format(curr)] + df['trade_in_1h_usdtfutures_{}_close_price'.format(curr)]
    df['final_trade_{}'.format(curr)] = df['trade_in_1h_usdtfutures_{}_close_price'.format(curr)]
    continue
    first_cols.append('final_trade_{}'.format(curr))
    cols_to_remove.append('trade_in_3h_usdtfutures_{}_close_price'.format(curr))
    cols_to_remove.append('trade_in_2h_usdtfutures_{}_close_price'.format(curr))
    cols_to_remove.append('trade_in_1h_usdtfutures_{}_close_price'.format(curr))
```

```python
for one in first_cols:
    cols = list(df)
    cols.insert(0, cols.pop(cols.index(one)))
    df = df.reindex(columns= cols)

df = df.drop(columns=cols_to_remove)
df
```

```python
final_trade_cols = []
for one in df.columns:
    if 'final_trade_' in one and PICKED_COIN in one:
        final_trade_cols.append(one)
final_trade_cols
```

```python
enc_features = rest_features
X = df[enc_features]
X
```

```python
Y = df[final_trade_cols]
Y = Y.eq(Y.where(df > 0).max(1), axis=0).astype(int)
Y
```

```python
from xgboost import XGBClassifier, XGBRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=6, shuffle=False)

train_test_split
```

```python
model = XGBClassifier()
model.fit(X_train, Y_train)
print(model)
```

```python
model.feature_importances_
```

```python
y_pred = model.predict(X_test)
```

```python
y_pred
predictions = [round(value) for value in y_pred]
```

```python
accuracy = accuracy_score(Y_test, predictions)
print("Accuracy: %.2f%%" % (accuracy * 100.0))
```

```python
def extract_validation_part(df):
    validation_size = int(len(df)/6)
    return df[-validation_size:]

def extract_validation_part_3hour(df):
    df = extract_validation_part(df)
    return df
    return df[::3]
```

```python
X_real = df[enc_features]

X_real = extract_validation_part_3hour(X_real)
X_real
```

```python
Y_pred_real = model.predict_proba(X_real)
#Y_pred_real = Y_pred_real *2
#Y_pred_real = Y_pred_real -1
Y_pred_real = [x[1] for x in Y_pred_real]
Y_pred_real
```

```python
Y_real = df[final_trade_cols]
Y_real = extract_validation_part_3hour(Y_real)
Y_real = Y_real.values
Y_real = [x[0] for x in Y_real]
assert len(Y_pred_real) == len(Y_real)
Y_real
```

```python
initial_bank = 400000
current_bank = initial_bank
rake = 0.00018
#rake = 0.00
hits = 0
shorts_hit = 0
shorts_miss = 0
longs_hit = 0
longs_miss = 0

last_side = 1
for i in range(len(Y_real)):
    side = 1 if Y_pred_real[i] >0.5 else -1
    leverage = 1
    if Y_pred_real[i] > 0.80 or Y_pred_real[i] < 0.20:
        leverage = 1
    profit = -1
    if Y_real[i] * side > 0:
        profit = 1
        hits +=1
        
    if side > 0 and leverage: #longs
        if Y_real[i] > 0:
            longs_hit+=1
        else:
            longs_miss+=1
    elif leverage: # shorts
        if Y_real[i] < 0:
            shorts_hit+=1
        else:
            shorts_miss+=1
            
            
    if last_side != side:
        # open rake
        current_bank = current_bank - (current_bank*rake*leverage)
    
    change = abs(Y_real[i]) * current_bank * profit* leverage
    current_bank = current_bank + change
    
    if last_side != side:
        # liq rake
        current_bank = current_bank - (current_bank*rake*leverage)
    last_side = side
    
    
current_bank, shorts_hit, shorts_miss, longs_hit, longs_miss, len(Y_real)
```
