```python
import pandas as pd
import pmdarima as pm
import tqdm
from statsmodels.tsa.arima.model import ARIMA

coin = "ADA"
df_orig = pd.read_hdf("v2_lagging_5jul_8may_2020_21_ada.hdf")

X = df_orig["usdtfutures_{}USDT_close_price".format(coin)].values
model_history_size = 60
forecast_minutes = 60


df_orig["c6"]=0


ORDER_REDO_COUNTER = 25
order_redo_counter = 1

for i in tqdm.tqdm(range(model_history_size, len(X) - forecast_minutes, 15)):
    order_redo_counter -=1
    x = X[i - model_history_size : i]
    if order_redo_counter ==0:
        order_redo_counter = ORDER_REDO_COUNTER
        
        model = pm.auto_arima(
            x,
            max_p=6,
            max_d=6,
            max_q=3,
            start_q=1,
            method="lbfgs",
            trace=False,
            alpha=0.07,
            information_criterion="aic",
            seasonal=False,
            stationary=False,
        )
        order = model.get_params()["order"]
        
    model = ARIMA(x, order=order)
    model_fit = model.fit()
    output = model_fit.forecast(steps=forecast_minutes)
    output_sum = sum(output)
    df_orig["c6"].iloc[i:i+forecast_minutes]=1 if output_sum > 0 else -1
```
