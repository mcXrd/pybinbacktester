```python
import pandas as pd
from xgboost import XGBClassifier, XGBRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import numpy as np
COIN ="ADA"
```

```python
#df_orig = pd.read_hdf("v2_lagging_2feb_8may_2021.hdf")
#df_orig = pd.read_hdf("v2_lagging_5jul_25dec_2020.hdf")
df_orig = pd.read_hdf("allfeatures.hdf")
df_orig = df_orig.fillna(0)
df_orig
```

```python
#sl = int(len(df_orig)/3)
df_orig
shift1 = 160
days = 35
start = -60*24*(days+shift1)
end = -60*24*shift1
df_orig = df_orig[start:end]
len(df_orig)/60/24
```

```python
df = df_orig.copy()
Y = df["trade_in_1h_spot_{}USDT_close_price".format(COIN)]
Y
```

```python
Y_MEAN = np.mean(np.abs(Y))
Y_MEAN = Y_MEAN
```

```python
Y = df["trade_in_1h_spot_{}USDT_close_price".format(COIN)]
Y[df["trade_in_1h_spot_{}USDT_close_price".format(COIN)]>Y_MEAN]=20
Y[df["trade_in_1h_spot_{}USDT_close_price".format(COIN)]<-Y_MEAN]=10
Y[df["trade_in_1h_spot_{}USDT_close_price".format(COIN)]<9]=1
Y
```

```python
unique, counts = np.unique(Y, return_counts=True)
unique, counts
```

```python
X = df_orig.loc[:,"spot_{}USDT_close_price".format(COIN):]
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=6, shuffle=False)
```

```python
model = XGBClassifier(objective='multi:softprob', num_class=3,booster="gbtree",
                      gamma=0.1,sampling_method=['uniform','gradient_based'][0],
                     num_parallel_tree=1, reg_alpha=1, reg_lambda=0, eta=0.3)
model.fit(X_train, Y_train)
print(model)
#model.feature_importances_
```

```python
y_pred = model.predict(X_test)
y_pred
unique, counts = np.unique(y_pred, return_counts=True)
unique, counts
```

```python
Y_TEST_REAL = df_orig["trade_in_1h_usdtfutures_{}USDT_close_price".format(COIN)]
Y_TEST_REAL = Y_TEST_REAL[-len(y_pred):]
Y_TEST_REAL
```

```python
assert len(Y_TEST_REAL) == len(Y_test)

for i in range(len(Y_TEST_REAL)):
    v = 1
    if Y_test[i] == 10:
        v=-1
    if Y_test[i] ==1:
        continue
    assert v * Y_TEST_REAL[i] > 0
    
    
```

```python
initial_bank = 400000
rake = 0.00018
```

```python
trading_hours = 0
skipped_hours = 0
last_side = -1
for i in range(1,len(Y_TEST_REAL),60):
    trading_hours+=1
    y_p = y_pred[i]
    y_r = Y_TEST_REAL[i]
    if y_p == 1:
        skipped_hours +=1
        continue
    
    side = -1
    if y_p == 20:
        side = 1
    
    profit = -1
    if y_r * side > 0:
        profit = 1
    
    change = initial_bank * abs(y_r) * profit
    initial_bank = initial_bank + change
    
    if side != last_side:
        rake_change = initial_bank * rake
        rake_change = rake_change * 2
        initial_bank = initial_bank - rake_change
    
    
initial_bank , trading_hours, skipped_hours/trading_hours
```
