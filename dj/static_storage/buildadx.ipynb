```python
df_orig["upmove"] = df_orig["usdtfutures_ADAUSDT_high_price"].rolling(60).mean() - df_orig["usdtfutures_ADAUSDT_high_price"].rolling(120).mean()

df_orig.upmove[df_orig["upmove"] < 0] = 0
```

```python
df_orig["downmove"] = df_orig["usdtfutures_ADAUSDT_low_price"].rolling(120).mean() - df_orig["usdtfutures_ADAUSDT_low_price"].rolling(60).mean()
df_orig.downmove[df_orig["downmove"] < 0] = 0
```

```python
df_orig["dmplus"] = 0
df_orig["dmplus"] = df_orig[df_orig["upmove"] > df_orig["downmove"]].upmove
df_orig["dmminus"] = 0
df_orig["dmminus"] = df_orig[df_orig["upmove"] < df_orig["downmove"]].downmove
df_orig = df_orig.fillna(0)
```

```python
df_orig["trend"]=0
df_orig.trend[df_orig["dmplus"] > 0] = 1
df_orig.trend[df_orig["dmminus"] > 0] = -1

```
